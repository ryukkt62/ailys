```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 1. Data Load

```python
data1 = pd.read_csv(r'Download dataset in TSV format-5.tsv', 
                    sep='\t',compression='gzip')
data2 = pd.read_csv(r'Download dataset in TSV format-2.tsv', 
                    sep='\t',compression='gzip')
```

```python
display(data1.info())
display(data1)
display(data2.info())
display(data2)
```

### Step 2. Data Preprocessing


#### 2.1. 컬럼 정리

```python
# data1
# 컬럼 분리
split_columns = data1.columns[0].split(',')  # 첫 번째 컬럼을 쉼표로 분리
split_columns[-1] = split_columns[-1].split('\\')[0]  # 마지막 부분의 '\TIME_PERIOD' 처리

# 원본 데이터의 첫 번째 컬럼을 쉼표로 분리하여 새로운 컬럼들 생성
new_columns = data1[data1.columns[0]].str.split(',', expand=True)
new_columns.columns = split_columns

# 나머지 컬럼들과 합치기
result1 = pd.concat([new_columns, data1.iloc[:, 1:]], axis=1)

# data2
# 컬럼 분리
split_columns = data2.columns[0].split(',')  # 첫 번째 컬럼을 쉼표로 분리
split_columns[-1] = split_columns[-1].split('\\')[0]  # 마지막 부분의 '\TIME_PERIOD' 처리

# 원본 데이터의 첫 번째 컬럼을 쉼표로 분리하여 새로운 컬럼들 생성
new_columns = data2[data2.columns[0]].str.split(',', expand=True)
new_columns.columns = split_columns

# 나머지 컬럼들과 합치기
result2 = pd.concat([new_columns, data2.iloc[:, 1:]], axis=1)
```

```python
display(result1)
display(result2)
```

#### 2.2. Missing Value 처리 및 필요한 값 추출


```python
data2.columns
```

```python
# 필요한 컬럼만 추출출
result11 = result1[['accomunit', 'unit', 'nace_r2', 'geo','2016 ']] # 2016 뒤에 빈칸 하나 들어가있음음
result11 = result11.loc[(result11['unit'] == 'NR') & (result11['accomunit'] == 'BEDPL') & 
                        (result11['nace_r2'] == 'I551')]


# 필요한 컬럼만 추출출
result22 = result2[['ind_type', 'indic_is', 'geo','unit','2016 ']] # 2016 뒤에 빈칸 하나 들어가있음음
result22 = result22.loc[(result22['ind_type'] == 'IND_TOTAL') & (result22['indic_is'] == 'I_IUG_TV') &
                        (result22['unit'] == 'PC_IND')]
```

```python
display(result11)
display(result22)
```

```python
# result11 처리
# 1. 2016 컬럼 처리
result11['2016 '] = result11['2016 '].replace(': ', np.nan)  # ':' 를 NaN으로 변경
result11['2016 '] = result11['2016 '].apply(lambda x: np.nan if isinstance(x, str) and ('u' in x or 'bu' in x or 'b' in x) else x)

# 2. geo 컬럼 처리
exclude_geo = ['EA', 'EU27_2007', 'EU27_2020', 'EU28']
result11 = result11[~result11['geo'].isin(exclude_geo)]

# 숫자형으로 변환 (공백 제거 후)
result11['2016 '] = result11['2016 '].str.strip().astype(float)
result11 = result11.dropna()



# result22 처리
# 1. 2016 컬럼 처리
result22['2016 '] = result22['2016 '].replace(': ', np.nan)  # ':' 를 NaN으로 변경
result22['2016 '] = result22['2016 '].apply(lambda x: np.nan if isinstance(x, str) and ('u' in x or 'bu' in x or 'b' in x) else x)

# 2. geo 컬럼 처리
result22 = result22[~result22['geo'].isin(exclude_geo)]

# 숫자형으로 변환 (공백 제거 후)
result22['2016 '] = result22['2016 '].str.strip().astype(float)
result22 = result22.dropna()






# 결과 확인
print("처리된 데이터 shape:")
print("result11:", result11.shape)
print("result22:", result22.shape)

# 결측치 확인
print("\n결측치 개수:")
print("result11:\n", result11.isna().sum())
print("\nresult22:\n", result22.isna().sum())
```

```python
# 나라 중복 확인인
print(result11.duplicated(subset='geo').shape)
print(result22.duplicated(subset='geo').shape)
```

#### 2.3. 데이터 병합

```python
# 컬럼 이름 변경
result11 = result11.rename(columns={'2016 ': 'Number of Bed-places',
                                    'geo':'Country Code'})
result22 = result22.rename(columns={'2016 ': 'Percentage of individuals',
                                    'geo':'Country Code'})
```

```python
total_data = result11[['Country Code','Number of Bed-places']].merge(result22[['Country Code','Percentage of individuals']], on='Country Code', how='inner').sort_values(by='Country Code', ascending=True)
total_data.shape
total_data.head()
total_data.tail()
total_data.info()
total_data.describe()

```

### Step 3. Visualization


* #### 높은 스마트 TV 사용률 + 많은 호텔 침대 공간: 제품을 출시하기에 가장 적합한 시장을 나타냄. 
* #### TV를 통한 인터넷 접속률이 높다는 것은 스마트 TV 기술에 익숙한 소비자 기반이 있다는 의미이고, 침대 공간이 많다는 것은 호텔 시장이 큰 것을 의미
* #### 시장 규모와 기술 수용도의 균형: 침대 공간이 많더라도 TV를 통한 인터넷 사용률이 낮다면 제품 수용이 느릴 수 있음
* #### 반대로, TV 사용률이 높더라도 호텔 시장이 작다면 총 시장 기회가 제한될 수 있음

#### 3.1. 산점도 시각화
* #### 침대 공간 수(X축)와 TV 인터넷 사용 비율(Y축)을 보여주는 산점도를 통해 두 지표가 모두 높은 국가를 쉽게 식별할 수 있음

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 산점도 생성
plt.figure(figsize=(12, 8))
sns.scatterplot(data=total_data, x='Number of Bed-places', y='Percentage of individuals')

# 국가 코드로 포인트 라벨링
for i, row in total_data.iterrows():
    plt.annotate(row['Country Code'], 
                 (row['Number of Bed-places'], row['Percentage of individuals']),
                 xytext=(5, 5), textcoords='offset points')

# 그래프 제목 및 축 라벨
plt.title('Hotel Bed Places vs TV Internet Usage by Country', fontsize=15)
plt.xlabel('Number of Hotel Bed Places', fontsize=12)
plt.ylabel('Percentage of Individuals Using Internet via TV (%)', fontsize=12)

# 그리드 추가
plt.grid(True, linestyle='--', alpha=0.7)

# 최적 시장 영역 표시 (선택 사항)
plt.axhspan(total_data['Percentage of individuals'].median(), 
            total_data['Percentage of individuals'].max(),
            total_data['Number of Bed-places'].median(),
            total_data['Number of Bed-places'].max(),
            alpha=0.2, color='green', label='Optimal Market')

plt.savefig('market_analysis_scatter.png')
plt.show()
```

#### 3.2. 복합 막대 그래프 (Dual Bar Chart)
* #### 각 국가별로 두 지표를 나란히 비교할 수 있는 막대 그래프


```python
# 데이터 정렬 (TV 사용률 기준)
sorted_data = total_data.sort_values(by='Percentage of individuals', ascending=False)

# 복합 막대 그래프 생성
fig, ax1 = plt.subplots(figsize=(14, 8))

# 첫 번째 축 (TV 사용률)
color = 'tab:blue'
ax1.set_xlabel('Country', fontsize=12)
ax1.set_ylabel('Percentage of Individuals (%)', color=color, fontsize=12)
ax1.bar(sorted_data['Country Code'], sorted_data['Percentage of individuals'], color=color, alpha=0.7, label='TV Internet Usage')
ax1.tick_params(axis='y', labelcolor=color)

# 두 번째 축 (침대 공간 수)
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Number of Bed Places', color=color, fontsize=12)
ax2.bar(sorted_data['Country Code'], sorted_data['Number of Bed-places'], color=color, alpha=0.3, label='Bed Places')
ax2.tick_params(axis='y', labelcolor=color)

# 그래프 제목
plt.title('TV Internet Usage and Hotel Capacity by Country', fontsize=15)

# 범례 추가
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('market_comparison_bars.png')
plt.show()
```

#### 3.3. 지표 결합 지수 시각화 (Market Potential Index)
* #### 두 지표를 결합하여 "시장 잠재력 지수"를 만들고 이를 시각화할 수 있음


```python
# 두 지표를 표준화 (0-1 범위로)
total_data['Standardized TV Usage'] = (total_data['Percentage of individuals'] - total_data['Percentage of individuals'].min()) / (total_data['Percentage of individuals'].max() - total_data['Percentage of individuals'].min())
total_data['Standardized Bed Places'] = (total_data['Number of Bed-places'] - total_data['Number of Bed-places'].min()) / (total_data['Number of Bed-places'].max() - total_data['Number of Bed-places'].min())

# 시장 잠재력 지수 계산 (두 요소의 가중 평균, 가중치는 필요에 따라 조정 가능)
total_data['Market Potential Index'] = (total_data['Standardized TV Usage'] * 0.5) + (total_data['Standardized Bed Places'] * 0.5)

# 결과 정렬
total_data_sorted = total_data.sort_values(by='Market Potential Index', ascending=False)

# 시각화
plt.figure(figsize=(12, 8))
bars = plt.bar(total_data_sorted['Country Code'], total_data_sorted['Market Potential Index'], color='teal')

# 상위 5개 국가 강조
for i, bar in enumerate(bars[:5]):
    bar.set_color('darkgreen')

plt.title('Market Potential Index for Smart Night Lamps by Country', fontsize=15)
plt.xlabel('Country Code', fontsize=12)
plt.ylabel('Market Potential Index (0-1)', fontsize=12)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 상위 5개 국가에 수치 표시
for i, v in enumerate(total_data_sorted['Market Potential Index'][:5]):
    plt.text(i, v + 0.02, f'{v:.2f}', ha='center')

plt.tight_layout()
plt.savefig('market_potential_index.png')
plt.show()
```
